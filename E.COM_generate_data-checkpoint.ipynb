## Creating a sample data or fake data with the help of python for Data analysis in E commerce

```python
# Importing useful libries
import pandas as pd 
import numpy as np
from faker import Faker
from tqdm import tqdm
import random
```

```python
# Initialize Faker 
fake = Faker()
print("🚀 Starting data generation process...")
```

```python
# ----------------------------------------------------
# 1. GENERATE DIM_PRODUCTS (1,000 Unique Products)
# ----------------------------------------------------
print("📦 Generating Product Dimension Table...")
categories = {
    'Electronics': ['Smartphones', 'Laptops', 'Headphones', 'Smartwatches', 'Keyboards'],
    'Home & Kitchen': ['Blenders', 'Coffee Makers', 'Air Fryers', 'Cookware Set', 'Toasters'],
    'Apparel': ['T-Shirts', 'Hoodies', 'Jeans', 'Sneakers', 'Running Shoes'],
    'Fitness': ['Yoga Mats', 'Dumbbells', 'Resistance Bands', 'Water Bottles', 'Treadmills']
}
```

```python
product_data = []
for p_id in range(1, 1001):
    category = random.choice(list(categories.keys()))
    sub_category = random.choice(categories[category])
    product_name = f"{fake.word().capitalize()} {sub_category[:-1] if sub_category.endswith('s') else sub_category}"
    # Base cost and a realistic retail price markup
    base_cost = round(random.uniform(5.0, 300.0), 2)
    unit_price = round(base_cost * random.uniform(1.3, 1.8), 2)  # 30% to 80% markup
    
    product_data.append([p_id, product_name, category, sub_category, base_cost, unit_price])

df_products = pd.DataFrame(product_data, columns=['Product_ID', 'Product_Name', 'Category', 'Sub_Category', 'Base_Cost', 'Unit_Price'])
df_products.head(6)
```

```python
# ----------------------------------------------------
# 2. GENERATE DIM_CUSTOMERS (50,000 Unique Customers)
# ----------------------------------------------------
print("👥 Generating Customer Dimension Table...")
customer_data = []
# Generating 50k profiles manually with a loop to ensure clean distinct fields
for c_id in tqdm(range(1, 50001)):
    customer_data.append([
        c_id,
        fake.name(),
        random.choice(['Male', 'Female', 'Other']),
        fake.city(),
        fake.state()
    ])
df_customers = pd.DataFrame(customer_data, columns=['Customer_ID', 'Customer_Name', 'Gender', 'City', 'State'])
df_customers.head(5)
```

```python
# ----------------------------------------------------
# 3. GENERATE DIM_CARRIERS (4 Shipping Options)
# ----------------------------------------------------
print("🚛 Generating Carrier Dimension Table...")
carrier_data = [
    ['CR-01', 'BlueDart', 'Express', 2],
    ['CR-02', 'Delhivery', 'Standard', 4],
    ['CR-03', 'FedEx', 'Express', 1],
    ['CR-04', 'DHL Logistics', 'Standard', 5]
]
df_carriers = pd.DataFrame(carrier_data, columns=['Carrier_ID', 'Carrier_Name', 'Ship_Mode', 'Avg_Delivery_Days'])
df_carriers.head(5)
```

```python
# ----------------------------------------------------
# 4. GENERATE FACT_ORDERS (1,000,000 Transactional Rows)
# ----------------------------------------------------
print("📊 Generating 1,000,000 Row Order Fact Table ")
num_rows = 1000000

# Using High-Performance NumPy vectors for IDs, quantities, and dates
order_ids = np.random.randint(10000000, 99999999, size=num_rows)
customer_ids = np.random.randint(1, 50001, size=num_rows)
product_ids = np.random.randint(1, 1001, size=num_rows)
carrier_ids = np.random.choice(['CR-01', 'CR-02', 'CR-03', 'CR-04'], size=num_rows, p=[0.4, 0.3, 0.15, 0.15])
quantities = np.random.randint(1, 6, size=num_rows) # Buy between 1 to 5 items

# Generate random dates over a 3-year timeline (2023 to 2025)
start_date = np.datetime64('2023-01-01')
end_date = np.datetime64('2026-01-01')
del_days = (end_date - start_date).astype(int)
random_days = np.random.randint(0, del_days, size=num_rows)
order_dates = start_date + random_days

# Shipping and Delivery Status logic
shipping_costs = np.round(np.random.uniform(3.0, 25.0, size=num_rows), 2)
order_statuses = np.random.choice(['Delivered', 'Returned', 'Delayed', 'Cancelled'], size=num_rows, p=[0.82, 0.08, 0.07, 0.03])

df_orders = pd.DataFrame({
    'Order_ID': order_ids,
    'Order_Date': order_dates,
    'Customer_ID': customer_ids,
    'Product_ID': product_ids,
    'Carrier_ID': carrier_ids,
    'Quantity': quantities,
    'Shipping_Cost': shipping_costs,
    'Order_Status': order_statuses
})
df_orders.size
```

```python
print("💾 Saving dataframes to CSV files...")
df_products.to_csv('dim_products.csv', index=False)
df_customers.to_csv('dim_customers.csv', index=False)
df_carriers.to_csv('dim_carriers.csv', index=False)
df_orders.to_csv('fact_orders.csv', index=False)

print("✅ Success! All 4 relational tables created successfully.")
```
